```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk import punkt
from nltk.corpus import stopwords
```

```python
ex= 'I am an Indian and also an Asian. India Subcontinent comes in the southern part of asia '
```

```python
token_ex= word_tokenize(ex)
token_ex
```

```python
stopword = stopwords.words('english')
stopword
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer(r'[a-zA-Z][^\s]*\b')
result = tokenizer.tokenize(ex)
```

```python
result
```

```python
tags = nltk.pos_tag(result , tagset = "") 
```

```python
tags
```

```python
from nltk.chunk import RegexpParser as rp
```

```python
chunkString = r''' Chunk: {<DT>?<JJ>*<NN>} '''
chunkParse = rp(chunkString)
chunkedwords = chunkParse.parse(tags)
```

```python
chunkedwords.draw()
```

```python
print('Chunks obtained: \n')
for subtree in chunkedwords.subtrees():
    if subtree.label() == 'Chunk':
        print(subtree)
```

```python
tokenizer1 = RegexpTokenizer(r'[a-zA-Z][^\s]*\b')
result = tokenizer1.tokenize(text1)
```

```python
result
```

```python
tags1 = nltk.pos_tag(result , tagset = "")
tags1
```

```python
chunkString = r''' Chunk: {<DT>?<JJ>*<NN>} '''
chinkstring = r''' Chink: {<JJ>*<NN>} '''
chinkParse = rp(chinkstring)
chunkParse = rp(chunkString)
chunkedwords = chunkParse.parse(tags1)
```

```python
chunkedwords.draw()
```

```python
print('Chunks obtained: \n')
for subtree in chunkedwords.subtrees():
    if subtree.label() == 'Chunk':
        print(subtree)
```

```python
chinkwords = chinkParse.parse(tags1)
```

```python
chinkwords.draw()
```

```python
print('Chinks obtained : \n ')
for subtree in chinkwords.subtrees():
    if subtree.label() == 'Chink':
        print(subtree)
```
